```python
import cv2
import matplotlib.pyplot as plt 
```

```python
img=cv2.imread("4.jpg",0)
plt.imshow(img,cmap="gray")
```

```python
# img=255-img
# plt.imshow(img,cmap="gray")
_,img1=cv2.threshold(img,100,255,cv2.THRESH_BINARY)
plt.imshow(img1,cmap="gray")
```

```python
contours,_=cv2.findContours(img1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
```

```python
#find contours
for c in contours :
    # print (len(c))
    if len(c)>1000 :
         # print (c)
            x,y,w,h=cv2.boundingRect(c)
            cv2.rectangle(img,(x,y),(x+w,y+h),(127,127,127),3)
plt.imshow(img,cmap="gray")          
```
