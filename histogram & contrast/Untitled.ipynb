```python
import matplotlib.pyplot as plt
import numpy as np 
import cv2
```

```python
img=cv2.imread("1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

hist=np.zeros((256),dtype=int)
rows,cols=img.shape

for i in range (rows) :
    for j in range (cols):
        index=img[i,j]
        hist[index] += 1
        
#print(hist)
plt.plot(hist)

    
```

```python
optimized_img = cv2.equalizeHist(img)
hist = cv2.calcHist([optimized_img], [0], None, [256],[0,256])

plt.plot(hist)

cv2.imshow('test',optimized_img)
cv2.imwrite('result.jpg',optimized_img)
cv2.waitKey(0)
```
