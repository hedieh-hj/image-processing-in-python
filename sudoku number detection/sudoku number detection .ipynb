```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img=cv2.imread ("sudoku.tif")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray") #inja bayd gray bezarim ke rangi nashe 
```

```python
img=cv2.equalizeHist(img)
plt.imshow(img,cmap="gray")
```

```python
rows , cols = img.shape
def  topleft():
    for i in range (rows):
        for j in range(cols):
            if img[i][j] != 255:
                #print("we couldnt find")
                
                return i,j
            
```

```python
def  downright():
    for i in range (rows-1,0,-1):   #baray be aghab omadn 
        for j in range(cols-1,0,-1):
            if img[i][j] != 255:
                #print("we couldnt find")
                
                return i,j
```

```python
x1,y1=topleft()  #top left 
x2,y2=downright()  #down right

print(x1,y1)
print(x2,y2)
```

```python
img=cv2.circle(img,(y1,x1),10,(0,0,0),-1)  #noghat goshe 
img=cv2.circle(img,(y2,x2),10,(0,0,0),-1)
img=cv2.circle(img,(y2,x1),10,(0,0,0),-1)
img=cv2.circle(img,(y1,x2),10,(0,0,0),-1)
plt.imshow(img,cmap="gray")
```

```python
img_finally=img[x1:x2 , y1:y2]
plt.imshow(img_finally,cmap="gray")
```

```python
# hala khone ha ro crop konim
w,h=img_finally.shape
print(w,h)
```

```python
def test(img):
    rows,cols =img.shape
    count=0
    for i in range (50,rows-50):
        for j in range (50,cols-50):
            if img [i][j] != 255:
                count+=1
    return count   
```

```python
w2=w//9
h2=h//9
counter=0
for i in range(0,w,w2):
    for j in range (0,h,h2):
        result_img=img_finally[i:i+w2,j:j+h2]
        if result_img.shape == (w2,h2):
            if test(result_img) > 10:
                cv2.imwrite(f"result/a{counter}.jpg",result_img)
                counter+=1
```
